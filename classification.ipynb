```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier

food_data = pd.read_csv('food_data.csv')
food_data.head()
food_data.info()
food_data.tail()
```

```python
# selecting the dependent and independent variables
food_data.index = food_data.iloc[:, 0]
x = food_data.iloc[:, 1:16]
y = food_data.iloc[:, 16]

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=420)

gnb = GaussianNB()
gnb.fit(x_train,y_train)
y_pred_gnb = gnb.predict(x_test)

db = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred_gnb})

db

metrics.accuracy_score(y_test, y_pred_gnb)
```

```python
candy_data = pd.read_csv('candy-data.csv')
candy_data.head()

# select the dependent and independent variables
x = candy_data.iloc[:, 10:13]
y = candy_data.iloc[:, 1]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 52)

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

log_reg = LogisticRegression()
log_reg.fit(x_train, y_train)
y_pred_log = log_reg.predict(x_test)

db = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred_log})

metrics.accuracy_score(y_test,y_pred_log)
```

```python
data = pd.read_csv('fruit_types.csv')

data.head()
```

```python
encoder = LabelEncoder()

fruit_name_encoded = encoder.fit_transform(data.iloc[:,0])
fruit_subtype_encoded = encoder.fit_transform(data.iloc[:,1])
mass_encoded = encoder.fit_transform(data.iloc[:,2])
width_encoded = encoder.fit_transform(data.iloc[:,3])
height_encoded = encoder.fit_transform(data.iloc[:,4])


```

```python
new_data = list(zip(fruit_name_encoded,fruit_subtype_encoded,mass_encoded,width_encoded,height_encoded))
new_data = pd.DataFrame(new_data, columns=['fruit_name', 'fruit_subtype', 'mass', 'width', 'height'])
data_to_use = new_data.iloc[:,1:5]
data_to_target = new_data.iloc[:,0]
new_data.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(data_to_use, data_to_target, test_size=0.2, random_state=123)
```

```python
knn_model = KNeighborsClassifier(n_neighbors=3) # define kNN model with k=3
knn_model.fit(x_train, y_train) # fit kNN model
y_pred = knn_model.predict(x_test)

print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```
