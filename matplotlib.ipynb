```python
import matplotlib.pyplot as plt
import numpy as np

y = np.linspace(-1,1,5)
y2 = np.linspace(-2,2,5)
x = np.arange(5)

# how to print figures
fig, ax = plt.subplots()
line1 = ax.plot(x,y)
line2 = ax.scatter(x,y)
line1 = ax.plot(x,y2)
line2 = ax.scatter(x,y2)

# figures with multiple plots
fig2, ax2 = plt.subplots(3,2)
ax2[0,1].plot(x,y)
ax2[0,0].plot(x,y2)
ax2[1,0].scatter(x,y2)

# how to close figures
plt.close(fig)
plt.close('all')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# different types of plot
y1 = np.arange(0,110,10)
y2 = np.random.random(11)
x = np.arange(11)

fig, ax = plt.subplots(2,2)
ax[0,0].plot(x,y1)
ax[0,1].scatter(x,y2)
ax[1,0].bar(x,y1)
ax[1,1].barh(x,y2)

# setting labels and titles
x = np.arange(0, 2*np.pi, 0.025)
ycos = np.cos(x)
ysin = np.sin(x)

fig2, ax2 = plt.subplots()
ax2.plot(x, ycos, 'b-', label = "cos(x)")
ax2.plot(x, ysin, 'r--', label = "sin(x)")
ax2.set_title("Trigonometric functions")
ax2.set_xlabel("0 to 2pi")
ax2.set_ylabel("cos(x) and sin(x)")
ax2.legend()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# vertical and horizontal lines

fig, ax = plt.subplots(1,2)

ax[0].axhline(0.3, color='red')
ax[0].axhline(0.4, linestyle = '--', color = 'blue')
ax[0].axhline(0.5, color='cyan', linewidth = 10)
ax[0].axhline(0.6, linestyle = ':', xmin=0.25, xmax = 0.75, color = 'green')
ax[0].axhline(0.7, xmin = 0.25, xmax = 0.75, color = (0.9, 0.2, 0.5, 0.6))

ax[1].axvline(0.3, color='red')
ax[1].axvline(0.4, linestyle = '--', color = 'blue')
ax[1].axvline(0.5, color='cyan', linewidth = 10)
ax[1].axvline(0.6, linestyle = ':', ymin=0.25, ymax = 0.75, color = 'green')
ax[1].axvline(0.7, ymin = 0.25, ymax = 0.75, color = (0.9, 0.2, 0.5, 0.6))

# set axes limits
ax[0].set_ylim([0.1,0.9])
ax[1].set_xlim([0.1,0.9])

# set title to figure and subplots
fig.suptitle("Horizontal and vertical lines")
ax[0].set_title("Horizontal lines", fontstyle='italic')
ax[1].set_title("Vertical lines", color='green', fontname='Arial')
```

```python
# scatter plots and 3d scatter plots

import matplotlib.pyplot as plt
import numpy as np

x = 10*np.random.random(12)
y = 10*np.random.random(12)
z = 10*np.random.random(12)

fig, ax = plt.subplots(1,3)
ax[0].scatter(x,y)
ax[0].set_title('default')

ax[1].scatter(x,y,50,marker='+')
ax[1].set_title('size=50, style=+')

crosses = ax[2].scatter(x, y, 200, marker='+', linewidth=3)
crosses.set_edgecolors('green')
crosses.set_linewidth(2)
bullets = ax[2].scatter(x, y, 50, marker ='o', color = 'black')
bullets.set_edgecolors('red')
bullets.set_linewidth(1.5)
ax[2].set_title('mixed')


# 3d scatter plots

fig3d = plt.figure()
ax3d = np.array([fig3d.add_subplot(1,3,1, projection='3d'),
                 fig3d.add_subplot(1,3,2, projection='3d'),
                 fig3d.add_subplot(1,3,3, projection='3d')])
ax3d[0].scatter(x,y,z)
ax3d[0].set_title('default')

ax3d[1].scatter(x,y,z,s=50,marker='+')
ax3d[1].set_title('size=50, style=+')

crosses = ax3d[2].scatter(x, y, z,s=200, marker='+', linewidth=3)
bullets = ax3d[2].scatter(x, y, z,s=50, marker ='o', color = 'black')
bullets.set_edgecolors('red')
bullets.set_linewidth(1.5)
ax3d[2].set_title('mixed')
```

```python
# log scales

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,10,0.01)
y = np.exp(x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_yscale('log')
ax.grid(which='both', linewidth=0.5, color=[.75,.75,.75,.5])
t = ax.text(7.5, y[1], 'semi-log paper', fontweight='bold')
```

```python
# bar charts

import matplotlib.pyplot as plt
import numpy as np

people = ["Student A", "Student B"]
studentA = np.array([90,50,80,40])
studentB = np.array([75,45,60,95])
x = np.arange(len(studentA))

fig, ax = plt.subplots(1,2)

ax[0].bar(x, studentA, width=0.3)
ax[1].bar(x, studentB, width=0.3, color=[1,0.647,0,1])

for i in range(len(ax)):
    ax[i].set_ylim(0,100)
    ax[i].set_title(people[i])
    #ax[i].set_xlabel("exercises")
    ax[i].set_ylabel("mark (%)")
    ax[i].set_xticks(range(len(studentA)))
    ax[i].set_xticklabels(["ex" + str(i+1) for i in range(len(studentA))])

fig.tight_layout()

# grouped bar chart

fig2, ax2 = plt.subplots()
width = 0.3
s1bars = ax2.bar(x - width/2, studentA, width, label = "Student A")
s2bars = ax2.bar(x + width/2, studentB, width, label = "Student B")
ax2.set_ylim(0,100)
#ax2.set_xlabel("exercises")
ax2.set_ylabel("mark (%)")
ax2.set_xticks(range(len(studentA)))
ax2.set_xticklabels(["ex" + str(i+1) for i in range(len(studentA))])
ax2.legend()
fig2.suptitle("Mark comparison")

for bar, score in zip(s1bars+s2bars, np.hstack((studentA,studentB))):
    if score < 50:
        bar.set_linewidth(3.5)
        bar.set_edgecolor('red')
```

```python
# drawing shapes

from matplotlib.patches import Polygon
import matplotlib.pyplot as plt
import numpy as np

# scaled plot so our circles show as circles and not ovals
fig, ax = plt.subplots()
ax.set_xlim([-10,10])
ax.set_ylim([0,20])
ax.axis('scaled')

# specs for rectangle 1
start_x = -2.5
start_y = 2
width = 5
height = 6
facecolor = 'gray'

rectangle = plt.Rectangle((start_x, start_y), width, height, fc=facecolor)
ax.add_patch(rectangle)

rectangle2 = plt.Rectangle((-0.5, 2), 1, 2, fc = 'brown')
ax.add_patch(rectangle2)

# spec for circle
radius = 0.5
circle = plt.Circle((-1,6),radius,fc='yellow',ec='black',lw=1)
circle2 = plt.Circle((1,6),radius,fc='yellow',ec='black',lw=1)
ax.add_patch(circle)
ax.add_patch(circle2)

# specs for polygon
start_xy = np.array([-2.5,8])
end_xy = np.array([2.5,8])
top_xy = np.array([0,10])

triangle = Polygon([start_xy, end_xy, top_xy], color = 'red')
ax.add_patch(triangle)
```
